# Test 2
Removing below certain decibel (quietness issue)

```python
for p in paths:

    ex, sr = librosa.load(p, sr=44100)

    S = np.abs(librosa.stft(ex))

    # max = S.max()

    S_dB = librosa.amplitude_to_db(S, ref=np.max)

    # u.show_spectrogram(S_dB)

    S_dB[S_dB<-40] = -80

    u.show_spectrogram(S_dB)

    S_c = librosa.griffinlim(librosa.db_to_amplitude(S_dB))

    # u.show_spectrogram_audiodata(ex)

    sf.write('test2' + p.split('/')[1].split('.')[0] + '.mp3', S_c, 44100, format='MP3')
```

# Test 3
"Fixing" quietness

```python
for p in paths:

    ex, sr = librosa.load(p, sr=44100)

    S = np.abs(librosa.stft(ex))

    max = S.max()

    S_dB = librosa.amplitude_to_db(S, ref=np.max)

    # u.show_spectrogram(S_dB)

    S_dB[S_dB<-40] = -80

    u.show_spectrogram(S_dB)

    S_c = librosa.griffinlim(librosa.db_to_amplitude(S_dB) * max)

    # u.show_spectrogram_audiodata(ex)

    sf.write('test3' + p.split('/')[1].split('.')[0] + '.mp3', S_c, 44100, format='MP3')
```

# Test 4
Trying to reduce the muffled effect

```python
for p in paths:

    ex, sr = librosa.load(p, sr=44100)

    S = np.abs(librosa.stft(ex))

    max = S.max()

    S_dB = librosa.amplitude_to_db(S, ref=np.max)

    # u.show_spectrogram(S_dB)

    S_dB[S_dB<-60] = -80

    u.show_spectrogram(S_dB)

    S_c = librosa.griffinlim(librosa.db_to_amplitude(S_dB) * max)

    # u.show_spectrogram_audiodata(ex)

    sf.write('test4' + p.split('/')[1].split('.')[0] + '.mp3', S_c, 44100, format='MP3')
```
